```python
import pandas as pd
import numpy as np
```

```python
#Saving the CSV file to a pandas dataframe
df = pd.read_csv('779212040_122018_1623_airline_delay_causes.csv')
df.head()
```

```python
#getting Details of the dataframe
df.info()
```

```python
#Change some  of the coulumn names to meaningful alternatives which I will be using later for visualization
df = df.rename(columns={
    "arr_flights":"arrived",
    "arr_del15":"delayed",
    " arr_delay":"minutes_delayed"
   
})
```

```python
df.head()
```

```python
#airport_name has the city, state and airport name all together, so we are splitting them 
df['city'], airports_name = df.airport_name.str.split(",").str
df['state'],df['airport'] = airports_name.str.split(":").str
```

```python
df.head()
```

```python
# Saving the dataframe to an xlsx file for the next step of visualization
df.to_excel('airline_delay_data_final.xlsx', index=False)
```
